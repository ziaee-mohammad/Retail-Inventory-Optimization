```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

random.seed(42)
np.random.seed(42)

# 1. Sales Table
sale_data = []
date_range = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")
product_ids = range(1, 101)
store_ids = range(1, 5)

for i in range(1, 2001):
    sale_data.append({
        'sale_id': i,
        'product_id': random.choice(product_ids),
        'date': random.choice(date_range),
        'quantity_sold': random.randint(1, 50),
        'store_id': random.choice(store_ids),
        'sales_amount': random.uniform(5.0, 500.0)
    })

sales_df = pd.DataFrame(sale_data)

# 2. Product Table
category_choices = ['Electronics', 'Clothing', 'Home Goods', 'Toys', 'Groceries']
product_data = []

for i in product_ids:
    product_data.append({
        'product_id': i,
        'product_name': f'Product_{i}',
        'category': random.choice(category_choices),
        'price_per_unit': round(random.uniform(5.0, 100.0), 2),
        'supplier_id': random.randint(1, 20),
        'reorder_point': random.randint(5, 20),
        'lead_time': random.randint(1, 10)
    })

product_df = pd.DataFrame(product_data)

# 3. Inventory Table
inventory_data = []
for i in range(1, 1001):
    last_restock_date = datetime(2023, random.randint(1, 12), random.randint(1, 28))
    inventory_data.append({
        'inventory_id': i,
        'product_id': random.choice(product_ids),
        'store_id': random.choice(store_ids),
        'current_stock': random.randint(0, 200),
        'last_restock_date': last_restock_date,
        'restock_quantity': random.randint(5, 50)
    })

inventory_df = pd.DataFrame(inventory_data)

# 4. Supplier Table
supplier_data = []
for i in range(1, 21):
    supplier_data.append({
        'supplier_id': i,
        'supplier_name': f'Supplier_{i}',
        'lead_time': random.randint(1, 10),
        'contact_info': f'contact_{i}@supplier.com'
    })

supplier_df = pd.DataFrame(supplier_data)

# 5. Store Table
store_data = [
    {'store_id': 1, 'store_name': 'All-In-One Mart ', 'location': 'Delhi-NCR'},
    {'store_id': 2, 'store_name': 'Ultimate Retail Hub', 'location': 'Pune'},
    {'store_id': 3, 'store_name': 'One-Stop Shop', 'location': 'Bengaluru'},
    {'store_id': 4, 'store_name': 'Everthing and More Store', 'location': 'Chennai'}
]

store_df = pd.DataFrame(store_data)

# Save all tables as CSV
sales_df.to_csv('sales.csv', index=False)
product_df.to_csv('product.csv', index=False)
inventory_df.to_csv('inventory.csv', index=False)
supplier_df.to_csv('supplier.csv', index=False)
store_df.to_csv('store.csv', index=False)

print("CSV files have been created.")
```
